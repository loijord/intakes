## Pastabos, kaip interpretuoti surinktą informaciją <a class="anchor" id="b13"></a> [<a class="fa fa-angle-double-left"></a>](#a)

Kaip matome pateiktoje maistinių medžiagų suvestinėje, dalis medžiagų, tokių kaip įvairūs riebalai, dubliuojasi. Taip pat maistines medžiagas reikėtų skirstyti į naudingas ir žalingas. Be to, angliavandenių atveju organizmui žalingas kalorijų perteklius atsiranda ne dėl bendro angliavandenių kiekio maiste, o dėl to, kad gliukozės kiekis kraujyje išauga per staiga ir nesuvartotas kalorijų perviršis ima kauptis. Vadovautis vien tiek riebalų, tiek angliavandenių kiekiu maiste yra apgaulinga.

### Rekomendacinės kalorijų paros normos <a class="anchor" id="b131"></a> [<a class="fa fa-angle-double-left"></a>](#a)
Nacionalinis Jungtinės Karalystės Sveikatos centras rekomenduoja, kad moterims reikia maždaug 2000 kcal išlaikyti sveikam kūno svoriui, o vyrams - apie 2500 kcal. 

### Riebalai  <a class="anchor" id="b132"></a> [<a class="fa fa-angle-double-left"></a>](#a)
Apytikslė riebalų kiekio skaičiavimo formulė yra:

```
FAT = FAMS (monounsaturated) + FAPU (polyunsaturated) + FASAT (saturated) + FATRN (trans)
```

Mityboje reikėtų kiek įmanoma labiau vengti trans ir saturated (sočiųjų) riebalų ir juos kiek įmanoma pakeisti su unsaturated (nesočiaisiais). Vitaminai A, D, E, K ir CHOLE (cholesterolis) taip pat priskiriami prie riebalų, tačiau šių medžiagų kiekiai nėra reikšmingi palyginus su FAMS, FAPU, FASAT ir FATRN.

Pabandžius panagrinėti įvairių maisto produktų sudėtis paaiškėjo, kad dalies (<10%) riebalų kilmė dažnai būna nežinoma.

````python
get_nutrition('100g butter ghee').reindex(['FAT', 'FAMS', 'FAPU', 'FASAT', 'FATRN', 'CHOLE'])
````

````python
def fat_results(meal):
    nutrition = get_nutrition(meal).reindex(['FAT', 'FAMS', 'FAPU', 'FASAT', 'FATRN', 'CHOLE'])
    fat_components = nutrition[['quantity']].reindex(['FAMS', 'FAPU', 'FASAT', 'FATRN']).sum(axis=0)
    fat_components += nutrition.loc['CHOLE', 'quantity']/1000
    fat_real = nutrition.loc['FAT', 'quantity']
    return f'{meal}: FAMS + FAPU + FASAT + FATRN + CHOLE = {round(fat_components.values[0], 2)} out of {fat_real} FAT'
    
print(fat_results('100g coconut milk'))
print(fat_results('100g chicken'))
print(fat_results('100g oil'))
print(fat_results('100g butter'))
print(fat_results('100g butter ghee'))
````

* Išvados. Nagrinėjant produktų maistingumo deklaracijas atkreipti dėmesį vien tik į FAT (riebalų) rodiklį yra apgaulinga, Reiktų atsižvelgti į atskiras riebalų rūšis: FAMS (mononesočiuosius), FAPU (polinesočiuosius), FASAT (sočiuosius), FATRN (trans) ir CHOLE (cholesterolį). Naudingais laikomi tik nesotieji riebalai, kitų reiktų vengti.

### Angliavandeniai, cukrus ir skaidulos  <a class="anchor" id="b133"></a> [<a class="fa fa-angle-double-left"></a>](#a)

Dauguma žmonių gyvena savo susikurtose iliuzijose ir tiki, kad visi angliavandeniai yra vienodi, tačiau, iš tikrųjų, tai labai toli nuo tiesos. Tarp jų yra akivaizdūs skirtumai, ypatingai, kai kalbame apie poveikį organizmui. Detaliau jie yra aprašomi [šiame straipsnelyje](https://energijosproduktai.lt/receptai/cukraus-ir-pridetinio-cukraus-skirtumai/). Čia aptarsime tik pačius angliavandenių žymėjimus (CHOCDF, CHOCDF.net, FIBTG, SUGAR), tokio žymėjimo motyvaciją ir kodėl šie rodikliai nėra pakankami aiškinantis poveikį organizmui.
Angliavandeniai yra klasifikuojami į:

* Cukrūs
  * Monosacharidai, tokie kaip gliukozė ir fruktozė
  * Disacharidai, tokie kaip sacharozė (stalo cukrus), maltozė ir laktozė (pieno cukrus). 
* Kompleksiniai angliavandeniai, kitaip vadinami polisacharidais. Jie apima krakmolą ir maistines skaidulas. Krakmolas yra gliukozės polimeras. Maistinės skaidulos yra dažniausiai nevirškinami kompleksiniai angliavandeniai, aptinkami augalų ląstelių sienelėse (celiuliozė, hemiceliuliozė ir pektinai) ir įvairios kitos medžiagos (dervos, glitėsiai ir dumblių polisacharidai).

Analizuojant angliavandenių energetinę vertę pagal bendrą susitarimą jie nėra detaliai tiriami. Nurodoma angliavandenių vertė paprastai yra angliavandeniai pagal skirtumą (CHOCDF), t.y. tai, kas lieka maisto produkte atmetus vandenį, baltymus, riebalus ir kepimo liekanas. Ši dalis apima cukrus, krakmolą, skaidulas ir nedidelius kiekius kitų organinių junginių.

Patyrinėjus maistines įvairių produktų sudėtis pasimatė toks dėsningumas:

```
CHOCDF (angliavandeniai pagal skirtumą) = CHOCDF.net (angliavandeniai, neto) + FIBTG (skaidulos)
```

````python
get_nutrition('100g baked chickpeas').loc[['CHOCDF', 'CHOCDF.net', 'FIBTG', 'SUGAR']] 
````

Šiame pavyzdyje matyti, kad 62.95g (CHOCDF) = 50.75g (CHOCDF.net) + 12.2 (FIBTG). Cukrūs (SUGAR) yra sudėtinė angliavandenių dalis, todėl priskirtina CHOCDF kategorijai. 

* Vietoje išvadų - papildymas. Cukrų ir kitų angliavandenių kiekis nėra pakankamas įvertis sveikai mitybai įvertinti. Agliavandeniai organizmui yra būtini, tačiau iš jų įsisavinamos kalorijos tampa perteklinėmis ir žalingomis, kai gliukozės kiekis kraujyje padidėja per stipriai arba būna padidėjęs per ilgą laiką. Norint pamatuoti, kiek būtent išaugs gliukozės kiekis kraujyje po tam tikro produkto suvartojimo, reikėtų atsižvelgti tiek į to produkto glikeminį indeksą (GI), tiek į glikeminę apkrovą (GA). 

  * Glikeminis indeksas (GI) yra skirtas nustatyti suvartojamų angliavandenių kokybei tam tikrame produkte, t.y. ar suvartojami angliavandeniai yra gerieji, ar blogieji (taip vadinamųjų tuščiųjų kalorijų šaltinis)
  * Glikeminė apkrova (GA) - tai išvestinis rodiklis iš GI ir suvartojamų angliavandenių kiekio. Jis skirtas įvertinti, kokios suvartojamų produktų porcijos būtų normos ribose.

### Alkoholis  <a class="anchor" id="b134"></a> [<a class="fa fa-angle-double-left"></a>](#a)

Alkoholiniuose gėrimuose ir kituose produktuose aptinkamo alkoholio (etanolio) kaloringumas yra daug didesnis nei angliavandenių, labiau panašesnis į riebalų (angliavandeniai: 100g - 400kcal, alkoholis: 100g - 700kcal, riebalai: 100g - 900kcal). Alkoholio kalorijos dažnai apibūdinamos kaip ,,tuščios kalorijos" kaip kai kurių angliavandenių ir riebalų atveju, nes įsisavinamos kaip perteklius šalia tų kalorijų, kurias organizmas gauna su maistu. Išgėrus alkoholio nutrūksta įprasti maistinių medžiagų, baltymų, angliavandenių ir riebalų, įsisavinimo procesai ir prioritetu tampa alkoholio kaip toksinės medžiagos pašalinimas iš organizmo.

Noras valgyti nesveiką maistą po išgėrimo yra nulemtas apetitą kontroliuojančių hormonų. Taip pat dėl papildomo organizmui atsiradusio streso sumažėja sąmoningumas, dėl ko atsiranda noras pasirinkti ne tokį sveiką maistą.

Jungtinėje Karalystėje įstatymiškai nėra privaloma skelbti didesnio kaip 1.2% alk. tūrio gėrimų kalorijų informacijos, tačiau tyrimai rodo, kad apie pusė UK gamintojų skelbia. Nuo 2022 metų Anglijoje restoranai, barai ir didesnės išsineštinio maisto vietos yra įpareigotos pateikti alkoholinių gėrimų kalorijas net ir tais atvejais, kai alkoholiniai gėrimai neviršija 1.2% alk. tūrio.

### Folatai  <a class="anchor" id="b135"></a> [<a class="fa fa-angle-double-left"></a>](#a)

Folate (FOLFD)/ vitamin B9 is naturally present in many foods. Folic acid (FOLAC) exists in foods fortified with it. It is mandatory in many countries. It is converted into active form of folate (FOLDFE) by the body. Unlike folate, not all of the folic acid you consume is converted into the active form of folate. In January 1998, the U.S. Food and Drug Administration required food manufacturers to add folic acid to foods commonly eaten, including breads, cereals, pasta, rice, and other grain products, to reduce the risk of neural tube defects. This program has helped to increase the average folic acid intake by about 100 mcg/day (out of recommended 240). The following relation holds: 

```
FOLDFE = 1.7 x FOLAC + FOLFD
```

### Gautos maistinės sudėties lentelės apibendrinimas  <a class="anchor" id="b136"></a> [<a class="fa fa-angle-double-left"></a>](#a)

Norint paruošti surinktus duomenis tolimesniam apdorojimui, siūlau kaip kurias perteklines grafas iš maistinės sudėties lentelės (žr.: [Maistinė sudėtis](#b122)) pašalinti, o likusius elementus klasifikuoti taip:

* **Klasifikacija pagal kiekį**: mikromedžiagos (vitaminai ir mineralai) ir makromedžiagos (baltymai, angliavandeniai, riebalai ir vanduo)
  * Mineralai: `['CA', 'FE', 'K', 'MG', 'P', 'ZN']`
  * Vitaminai: `['FOLDFE', 'NIA', 'RIBF', 'THIA', 'TOCPHA', 'VITA_RAE', 'VITB12', 'VITB6A', 'VITC', 'VITD', 'VITK1']`
  * Mikromedžiagos = mineralai + vitaminai
  * Makromedžiagos: `['CHOCDF', 'PROCNT', 'FAT', 'WATER']`
* **Makromedžiagų klasifikacija pagal gerąsias ir blogąsias elementų atmainas**. Esmė: pateiktų medžiagų suvartojimas yra beveik neišvengiamas, tačiau reikėtų bloguosius elementus pakeisti geraisiais. Poreikis skirstyti atsiranda dažniausiai angliavandenių ir riebalų atveju. Priminsime, kad pagrindinės riebalų (FAT) sudedamosios dalys yra CHOLE, FAMS, FAPU, FASAT ir FATRN, o angliavandenių (CHOCDF) - cukrūs (SUGAR), skaidulos (FIBTG) ir kiti elementai. Tad belieka tik šiek tiek detaliau išskirti kai kurių makromedžiagų (`['CHOCDF', 'PROCNT', 'FAT', 'WATER']`) tipus. Taip pat įtraukta ir valgomoji druska (NA). Jos perviršis taip pat yra kenksmingas.
  * Blogosios medžiagos: `['CHOLE', 'FASAT', 'FATRN', 'NA', 'SUGAR']`
  * Gerosios medžiagos: `['FAMS', 'FAPU', 'FIBTG', 'WATER']`

````python
minerals = ['CA', 'FE', 'K', 'MG', 'P', 'ZN']
vitamins = ['FOLDFE', 'NIA', 'RIBF', 'THIA', 'TOCPHA', 'VITA_RAE', 'VITB12', 'VITB6A', 'VITC', 'VITD', 'VITK1']
macronutrients = ['CHOCDF', 'PROCNT', 'FAT', 'WATER']
micronutrients = minerals + vitamins

negative_nutrients = ['CHOLE', 'FASAT', 'FATRN', 'NA', 'SUGAR'] #limit your saturated fat, salt and sugar
positive_nutrients = ['FAMS', 'FAPU', 'FIBTG', 'WATER'] #consume more unsaturated fat and fiber
nutrients = negative_nutrients + positive_nutrients
````

## Mikromedžiagų kiekiai recepte, išreikšti dienos normomis <a class="anchor" id="b21"></a>  [<a class="fa fa-angle-double-left"></a>](#a)
Taip pat kiekvienai medžiagai pateikiamas pasiskirstymas pagal ingredientus

````python
import plotly.express as px

def fig_save(fig, path, width=850, height=600):
    fig.update_layout(width=width, height=height)
    with open(path, "w", encoding="utf-8") as f: 
        f.write(fig.to_html())

def plot_nutrient_distribution(nutrients, nutr_df, nutr_df_meta, intake_df):
    df = nutr_df.loc[nutrients].div(intake_df.loc[nutrients], axis=0)
    fig = px.bar(df, x=df.index, y=df.columns.tolist(),
                 labels={'index': 'Nutrients', 'value':'Daily norm', 'variable':'Ingredients'})
    fig.add_hline(y=1, line_width=1, line_dash="dash", line_color="green")
    fig.add_hline(y=d['Energy, kcal']/2500, line_width=1, line_dash="dot", line_color="green")
    fig.update_xaxes(
        ticktext=nutr_df_meta['label'].loc[nutrients],
        tickvals=df.index,
    )
    return fig
    
fig = plot_nutrient_distribution(micronutrients, nutr_df, nutr_df_meta, intake_df)
fig_save(fig, 'research_examples/sample1.html')
````

## Kiekvieno ingrediento naudingumas, išreikštas mikronutrientų dienos normomis <a class="anchor" id="b23"></a>  [<a class="fa fa-angle-double-left"></a>](#a)
Taip pat pateikiamas mikronutrientų, įeinančių į tas dienos normas, pasiskirstymas

````python
def plot_nutritional_scores(nutrients, nutr_df, nutr_df_meta, intake_df):
    df2 = nutr_df.loc[nutrients].div(intake_df.loc[nutrients], axis=0).T
    fig = px.bar(df2, y=df2.columns.tolist(), x=df2.index,
                 labels={'index': 'Ingredients', 'value':'Daily norms', 'variable':'Ingredients'})

    for idx, (n1, n2) in enumerate(zip(nutr_df_meta.loc[nutrients].index, 
                                       nutr_df_meta['label'].loc[nutrients])):
        fig.data[idx].name = n2
        fig.data[idx].hovertemplate = n2

    return fig
    
fig = plot_nutritional_scores(micronutrients, nutr_df, nutr_df_meta, intake_df)
fig_save(fig, 'research_examples/sample3.html')
````

## Kiekvieno ingrediento informacija <a class="anchor" id="b24"></a>  [<a class="fa fa-angle-double-left"></a>](#a)
Galima paskaičiuoti įvairius rodiklius apie kiekvieną ingredientą
* `price` - savikaina
* `weight` - svoris
* `microscore` - kiekvieno mikronutriento dienos normų suminė vertė
* `negscore` - kiekvienos blogosios medžiagos dienos limitų suminė vertė
* `posscore` - kiekvienos gerosios medžiagos dienos normų suminė vertė
* `microscore_per_price`, `negscore_per_price`, `posscore_per_price` - atitinkamų įverčių ir kainos santykis

Pastaba: baltymai nėra priskirtini prie gerųjų ar blogųjų medžiagų, nes į tyrimą nebuvo įtraukti, todėl reali situacija gali skirtis

````python
#| hide
df = price_df.T.reset_index().rename(columns={"index": "name", 0: "price"})
df['weight'] = nutr_df.loc['_WEIGHT'].values.round(3)
df['microscore'] = nutr_df.loc[micronutrients].div(intake_df.loc[micronutrients], axis=0).sum(axis=0).values.round(4)
df['negscore'] = nutr_df.loc[negative_nutrients].div(intake_df.loc[negative_nutrients], axis=0).sum(axis=0).values.round(4)
df['posscore'] = nutr_df.loc[positive_nutrients].div(intake_df.loc[positive_nutrients], axis=0).sum(axis=0).values.round(4)
df['microscore_per_price'] = df['microscore']/df['price']
df['negscore_per_price'] = df['negscore']/df['price']
df['posscore_per_price'] = df['posscore']/df['price']
df
````
